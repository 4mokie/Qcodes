```python
from threading import Thread
from queue import Queue
from itertools import count
```

```python
class Listener(Thread):
    def __init__(self, shape, q):
        self.c = count(0)
        self.q = q
        self.z_arr = np.ma.masked_invalid(np.empty(shape)* np.nan)
        self.stop = False
        self.fig, self.ax = plt.subplots()
        super().__init__()
    
    def run(self):
        while True:
            msg = self.q.get()
            if msg =="DONE":
                break
            else:
                print(msg)
                self.z_arr.flat[msg[0]-1] = msg[1]
            
            if next(self.c) % 200 == 0:
                self.ax.pcolormesh(self.z_arr)
                self.fig.canvas.draw()
```
